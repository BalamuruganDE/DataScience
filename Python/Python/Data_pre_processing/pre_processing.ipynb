```python
def dummy_df(df, list1):
    for x in list1:
        print(x,type(x))
        x=x.strip()
        dummies = pd.get_dummies(df[x], prefix = x, dummy_na = False,dtype=int)
        # dummies = pd.get_dummies(df[x],  dummy_na = False,dtype=int)
        df = df.drop(x, axis = 1)
        df = pd.concat([df, dummies], axis = 1)
    return df
```
