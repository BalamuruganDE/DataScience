```python
%%time
a=[[1,1.1,1.11,1.111],[2,2.2,2.22,2.222],[3,3.3,3.33,3.333],[4,4.4,4.44,4.444]]
print(a)
print(type(a))
m=n=0
for i in a:
    print('i=',i)
    for j in i:
        print('j=',j)
        print('m=',m,'n=',n)
        print('Before update: a[',m,'][',n,'] = ',a[m][n])
        a[m][n]=2* a[m][n]
        print('AFter update : a[',m,'][',n,'] = ',a[m][n])
        n+=1
    print('updated a[',m,']:',a[m])
    m+=1
    n=0
    
print(a)
```
