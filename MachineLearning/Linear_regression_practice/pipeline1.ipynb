```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest,f_regression

from sklearn.pipeline import Pipeline 
from sklearn.impute import SimpleImputer

from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression


import joblib
import os
```

```python
data = pd.read_csv('../LinearRegression/Dataset/housing_prices_dataset.csv')
```

```python
os.getcwd()
```

```python
data.head()
```

```python
def preprocess_data (data):
    data['HouseAge'] = 2024 - data['YearBuilt']
    data.drop(columns='YearBuilt',inplace=True)
    mapping = {'Low':1, 'Medium':2, 'High':3}
    data['LuxuryRating']=data['LuxuryRating'].map(mapping)
    data = pd.get_dummies(data,dtype='int')
    return data
```

```python
data
```

```python
data = preprocess_data(data)
```

```python
data
```

```python
data_temporary_just_time_being = data.dropna()
```

```python
data_temporary_just_time_being
```

```python
X=data_temporary_just_time_being.drop(columns='Price')
y=data_temporary_just_time_being['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
X = X.dropna()
```

```python
X
```

```python
select_K_best = SelectKBest(f_regression,k=5)
X_select = select_K_best.fit(X,y)
select_K_best.get_support()
selected_feature_names = X.columns[select_K_best.get_support()]
selected_feature_names
```

```python
selected_feature_names.values
```

```python
features = selected_feature_names.values
```

```python
transformer = Pipeline( steps=[('imputer',SimpleImputer(strategy='median')),
                               ('scaling',StandardScaler())]
)
```

```python
preprocessor = ColumnTransformer(transformers=[(('num',transformer,features))])
```

```python
pipeline = Pipeline(steps=[('preprocessor',preprocessor),('model',LinearRegression())])
```

```python
pipeline
```

```python
data = pd.read_csv('../LinearRegression/Dataset/housing_prices_dataset.csv')
data = preprocess_data(data)
X = data[features]
y = data['Price']
```

```python
X
```

```python
y
```

```python
pipeline.fit(X,y)
```

```python
predict_data_X = pd.DataFrame([[1700, 3, 1, 2, 30]], columns=features)
```

```python
pipeline.predict(predict_data_X)
```

```python
pipeline.coeff_
```
